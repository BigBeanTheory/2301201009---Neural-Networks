#1. Explain the role of weights in a neuron.

---

In an artificial neuron, weights determine how important each input is to the final output. Every input value is multiplied by a corresponding weight before being summed together. If a weight is large, that input has a strong influence on the neuron’s output. If a weight is small or close to zero, the input has little effect. During training, these weights are adjusted so the model can learn patterns from data and make accurate predictions.


#2. What is an activation function?


---

An activation function is a mathematical function applied to the weighted sum of inputs in a neuron to produce the final output. It introduces non-linearity into the model, allowing neural networks to learn complex patterns instead of just linear relationships. Without activation functions, a neural network would behave like a simple linear model regardless of its depth. Common examples include Sigmoid, ReLU, and Tanh.

#3. Define probability distribution in ML context.

---

In machine learning, a probability distribution describes how likely different outcomes or values are for a random variable. It helps model uncertainty in data and is used in tasks such as classification, regression, and generative modeling. For example, in classification, the model may output a probability distribution over possible classes, showing how confident it is about each class.

#4. What is gradient in optimization?

---

In optimization, the gradient is a vector of partial derivatives that shows the direction and rate of fastest increase of a function. In machine learning, it is commonly used to minimize a loss function. Algorithms like gradient descent use the gradient to update model parameters step by step in the direction that reduces error, helping the model improve its performance during training.
