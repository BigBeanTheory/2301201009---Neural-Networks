#1. What are epoch, batch size, and Stratified? Please explain them.
---

- Epoch: One epoch means the neural network has seen the entire training dataset exactly once. During training, the model processes all samples, computes the loss, and updates its weights (via backpropagation). Training usually runs for multiple epochs (e.g., 10, 50, 100) until the model converges or starts overfitting. More epochs means more learning opportunities, but too many can cause overfitting.

- Batch Size: Instead of updating weights after every single sample (batch size = 1) or after the whole dataset (batch size = full dataset), we divide the data into smaller batches.

  Small batch size means noisier but faster updates, better generalization.
  
  Large batch size means smoother gradients, more memory needed, sometimes stuck in sharp minima.

- Stratified: This refers to stratified sampling. It ensures that when we split data into train/validation/test sets (or folds), the class distribution remains the same as the original dataset.
Crucial for imbalanced datasets (e.g., 95% class A, 5% class B). Without stratification, one fold might accidentally get almost no minority class samples, leading to misleading evaluation.

#2. Explain the different types for activation functions?
---


Activation functions introduce non-linearity into neural networks, allowing them to learn complex patterns.

| Activation     | Range              | Common Use                              |
|----------------|--------------------|-----------------------------------------|
| **Sigmoid**    | (0, 1)             | Binary classification (output layer)    |
| **Tanh**       | (-1, 1)            | Hidden layers (older networks)          |
| **ReLU**       | [0, ∞)             | Default choice for hidden layers        |
| **Leaky ReLU** | (-∞, ∞)            | When ReLU causes dead neurons           |
| **ELU**        | (-α, ∞)            | Deeper networks                         |
| **Softmax**    | (0, 1), sums to 1  | Multi-class classification              |

#3. How do we handle structured and unstructured data in ANNs and CNNs?
---


**Structured Data** (tabular data):
- Examples: CSV files, spreadsheets, features like age, income, credit score, etc.
- Best Network: ANN (Artificial Neural Network / MLP)  
- How it is handled:  
  Each feature becomes one input neuron. The data is treated as a flat vector.  
  Fully connected (dense) layers are used. No spatial relationship is assumed between features.

**Unstructured Data** (images, text, audio, video):
- Examples: Photos, videos, raw text, sound waves.
- Best Network: CNN (Convolutional Neural Network) for image and video data.
- How it is handled:  
  Images are fed as 3D tensors (height × width × channels).  
  CNNs use convolution operations to automatically extract spatial hierarchies, starting from simple edges to complex objects.  
  This approach is much more efficient than using plain ANN for images because it uses parameter sharing and respects spatial relationships.

#4. What is a layer? Define ANN and CNN layers?
---

A layer is a group of neurons (or operations) that transform input into output. Each layer applies weights, bias, and usually an activation function.

ANN (Feedforward / MLP) Layers:

- Input Layer: One neuron per feature (e.g., 784 neurons for 28×28 flattened image).
- Hidden Layers (Dense / Fully Connected): Every neuron connected to every neuron in the previous layer. Typical: 1–5 hidden layers.
- Output Layer: 1 neuron (binary) or number of classes (multi-class) with softmax/sigmoid.


CNN Layers (in order):

- Convolutional Layer (Conv2D): Learns filters/kernels that slide over the image and detect features.
- Pooling Layer (MaxPool2D / AvgPool2D): Down-samples feature maps → reduces computation and adds translation invariance.
- Batch Normalization (optional but common): Stabilizes and speeds up training.
- Dropout (optional): Randomly drops neurons during training to prevent overfitting.
- Flatten: Converts 2D/3D feature maps into 1D vector.
- Fully Connected (Dense) Layers: Same as ANN hidden layers.

#5. What concepts are used in ANN and CNN?
---

Common to Both:

- Forward Propagation: Input → layers → prediction.
- Loss Function (e.g., Cross-Entropy, MSE).
- Backpropagation: Computes gradients of loss w.r.t. weights.
- Optimizer (Adam, SGD, RMSprop) + Learning Rate.
- Regularization: Dropout, L1/L2 weight decay, early stopping.
- Overfitting / Underfitting diagnosis via train/validation curves.

Specific to CNN:

- Convolution Operation: $  \text{Output}[i,j] = \sum \text{Input}[i+k, j+l] \cdot \text{Kernel}[k,l]  $
- Kernel / Filter: Small learnable matrix (3×3, 5×5, etc.).
- Stride & Padding: Controls how the filter moves and handles borders.
- Feature Maps: Output of each convolutional layer (channels = number of filters).
- Parameter Sharing & Sparse Connectivity: Makes CNNs efficient for images.
