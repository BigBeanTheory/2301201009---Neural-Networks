#Logistic Regression with Tensors

---



```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
hours = tf.constant([4.0, 2.3, 1.1, 5.4, 3.2, 6.7, 4.1, 0.4, 7.2, 2.9, 5.9, 1.8, 6.1, 3.7, 4.8], dtype=tf.float32)
hours = tf.reshape(hours, (-1, 1))

pass_exam = tf.constant([True, False, False, True, False, True, True, False, True, False, True, False, True, True, True], dtype=tf.bool)
```

```python
plt.scatter(hours, pass_exam)
plt.xlabel("Hours Studied")
plt.ylabel("Pass (0 or 1)")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(hours.numpy(), pass_exam.numpy(), test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.predict(np.array([1.6,1.5,0.3,7.0,3.5]).reshape(-1, 1))
```

```python
X_sorted = np.sort(hours, axis=0)
plt.plot(X_sorted, model.predict_proba(X_sorted)[:, 1], color='red')
plt.scatter(hours, pass_exam)
plt.xlabel("Hours Studied")
plt.ylabel("Probability of Passing")
plt.title("Logistic Regression Fit")
plt.show()
```
